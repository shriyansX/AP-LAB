```python
s = input("Enter a string: ")

rev = ""
for char in s:
    rev = char + rev 

print("Reversed string:", rev)
```

```python
s = input("Enter a string: ")

vowels = "aeiouAEIOU"
v_count = c_count = d_count = space_count = 0

for ch in s:
    if ch in vowels:
        v_count += 1
    elif ch.isalpha():
        c_count += 1
    elif ch.isdigit():
        d_count += 1
    elif ch.isspace():
        space_count += 1

print("Vowels:", v_count)
print("Consonants:", c_count)
print("Digits:", d_count)
print("Spaces:", space_count)
```

```python
s = input("Enter a string: ")

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```
