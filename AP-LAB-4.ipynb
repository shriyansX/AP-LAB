```python
lst = [10, 5, 20, 8, 15]

max_val = lst[0]
min_val = lst[0]

for i in lst:
    if i > max_val:
        max_val = i
    if i < min_val:
        min_val = i

print("Maximum:", max_val)
print("Minimum:", min_val)
```

```python
lst = [1, 2, 3, 2, 4, 2, 5]
element = 2
count = 0

for i in lst:
    if i == element:
        count += 1

print("Occurrences of", element, ":", count)
```

```python
lst = [1, 2, 3, 4, 5]
rev = lst[::-1]

print("Reversed List:", rev)
```

```python
lst = [1, 2, 2, 3, 4, 4, 5]
unique = []

for i in lst:
    if i not in unique:
        unique.append(i)

print("List without duplicates:", unique)
```

```python
lst = [5, 2, 9, 1, 7]

asc = sorted(lst)
desc = sorted(lst, reverse=True)

print("Ascending:", asc)
print("Descending:", desc)
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

merged = []
for i in list1:
    merged.append(i)
for j in list2:
    merged.append(j)

print("Merged List:", merged)
```

```python
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
common = []

for i in list1:
    if i in list2 and i not in common:
        common.append(i)

print("Common Elements:", common)
```

```python
lst = [1, 2, 3, 4, 5]
n = 2

left_rotate = lst[n:] + lst[:n]
right_rotate = lst[-n:] + lst[:-n]

print("Left Rotation:", left_rotate)
print("Right Rotation:", right_rotate)
```

```python
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

add = [[0, 0], [0, 0]]
sub = [[0, 0], [0, 0]]

for i in range(len(A)):
    for j in range(len(A[0])):
        add[i][j] = A[i][j] + B[i][j]
        sub[i][j] = A[i][j] - B[i][j]

print("Addition:", add)
print("Subtraction:", sub)
```
